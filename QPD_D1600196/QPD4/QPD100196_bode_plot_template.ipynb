```python
import numpy as np 
import matplotlib.pyplot as plt
import sdfascii as sdf
```

```python
COLOR = 'green'
plt.rcParams['text.color'] = COLOR
plt.rcParams['axes.labelcolor'] = COLOR
plt.rcParams['xtick.color'] = COLOR
plt.rcParams['ytick.color'] = COLOR
```

### Import test point 4 amplitude

```python
ch1_4 = sdf.read_ascii_files('CH1/414A')
ch2_4 = sdf.read_ascii_files('CH2/424A')
ch3_4 = sdf.read_ascii_files('CH3/434A')
ch4_4 = sdf.read_ascii_files('CH4/444A')
```

### Import test point 4 phase

```python
ch1_4p = sdf.read_ascii_files('CH1/414PH')
ch2_4p = sdf.read_ascii_files('CH2/424PH')
ch3_4p = sdf.read_ascii_files('CH3/434PH')
ch4_4p = sdf.read_ascii_files('CH4/444PH')
```

## Test point 4 bode

```python
fig, axs = plt.subplots(2, figsize=(10,10))
axs[0].semilogx(ch1_4.frequency[:], ch1_4.amplitude[:],color='red',label='CH1')
axs[0].semilogx(ch2_4.frequency[:], ch2_4.amplitude[:],color='blue', label='CH2')
axs[0].semilogx(ch3_4.frequency[:], ch3_4.amplitude[:], color='orange', label='CH3')
axs[0].semilogx(ch4_4.frequency[:], ch4_4.amplitude[:], color='magenta', label='CH4')
axs[0].title.set_text('QPD4 TP4')
axs[0].title.set_size(20)
axs[0].legend()

axs[0].set_ylabel('amplitude [dB]')
axs[1].semilogx(ch1_4p.frequency[:], ch1_4p.amplitude[:],color='red', label='CH1')
axs[1].semilogx(ch2_4p.frequency[:], ch2_4p.amplitude[:],color='blue',label='CH2')
axs[1].semilogx(ch3_4p.frequency[:], ch3_4p.amplitude[:],color='orange',label='CH3')
axs[1].semilogx(ch4_4p.frequency[:], ch4_4p.amplitude[:], color='magenta',label='CH2 V-')
axs[1].legend()
axs[1].set_ylabel('phase [deg]')
axs[1].set_xlabel('frequency [Hz]')
```

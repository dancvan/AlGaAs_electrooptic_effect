```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
folder = '../../../measurements/HVA_tf/'
```

```python
label = 'HVA_11_12_2019_'
```

```python
amp_1 = np.loadtxt(folder + label + 'amp.TXT') 
phase_1 = np.loadtxt(folder + label + 'phase.TXT')
amp_2 = np.loadtxt(folder + label + 'amp_half_knob.TXT') 
phase_2 = np.loadtxt(folder + label + 'phase_half_knob.TXT')
```

```python
fig = plt.figure(figsize=(17,15))
plt.subplot(211)
plt.title('High voltage amplifier (full amplitude) with pomona box (see alog 412)',fontsize=25)
ax = plt.gca()
plt.ylabel('dB',fontsize=20)
plt.grid('True', which='both')
plt.grid('True')
ax.tick_params(which='both', width=1)
ax.tick_params(which='minor', length=2)
ax.tick_params(labelsize=15)
ax.yaxis.offsetText.set_fontsize(15)
ax.axhline(amp_1[0,1],color='red',label=str(max(amp_1[:,1])) +' dB')
ax.axhline(amp_2[1,1],color='purple',label=str(max(amp_2[:,1])) +' dB')
plt.plot(amp_1[:,0],amp_1[:,1],label='max amplitude')
plt.plot(amp_2[:,0],amp_2[:,1], label = 'half amplitude')
plt.xlim([1,amp_1[-1,0]])
ax.set_xscale('log')
plt.legend(prop={'size':20})
plt.subplot(212)
plt.plot(phase_1[:,0],phase_1[:,1],label = 'max amplitude')
plt.plot(phase_2[:,0],phase_2[:,1], label = 'half ampltiude')
ax = plt.gca()
plt.ylabel('phase (degrees)',fontsize=20)
plt.grid('True', which='both')
plt.grid('True')
ax.tick_params(which='both', width=1)
ax.tick_params(which='minor', length=2)
ax.tick_params(labelsize=15)
ax.yaxis.offsetText.set_fontsize(15)
plt.xlabel('frequency [Hz]',fontsize=20)
plt.xlim([1,phase_1[-1,0]])
ax.set_xscale('log')
plt.legend(prop={'size':20}, loc=3)
plt.savefig(folder + 'HVA_plus_pomona_tf.png',dpi=300,bbox_inches='tight')
```

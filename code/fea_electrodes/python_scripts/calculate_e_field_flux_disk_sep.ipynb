```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import h5py
```

```python
params = {"ytick.color" : "g",
          "xtick.color" : "g",
          "axes.labelcolor" : "g",
          "axes.edgecolor" : "g",
          "text.color" : "g",
         }
plt.rcParams.update(params)
```

```python
file = h5py.File('data/split_electrode/numerical_split_electrode_test_vary_sep.h5', 'r')
file.keys()
```

```python
V_o = file['V_o']
lenX = file['lenX'][()]
lenZ = file['lenZ'][()]
X = file['X'][:,:]
Z = file['Z'][:,:]
U = file['potential'][:,:,:]
disk_radii = file['disk_radii'][()]
hole_radii = file['hole_radii'][()]
disk_separation = file['disk_separation'][:]
delta = file['delta'][()]
```

```python
E_U = np.empty((int(lenZ/delta)+1,int(lenX/delta),len(disk_separation)))
E_V = np.empty((int(lenZ/delta)+1,int(lenX/delta),len(disk_separation)))
Flux_rad = 500e-6
E_flux_rad = np.empty((int(Flux_rad/delta),len(disk_separation)))
Ez_flux = np.empty((len(disk_separation)))
Ez_center = np.empty((len(disk_separation)))
x_center = len(X[:,0])/2

for i in range(0, len(disk_separation)):
    E_U[:,:,i], E_V[:,:,i] = np.gradient(U[:,:,i], delta)
    E_U[:,:,i] = -1.0*E_U[:,:,i]
    E_V[:,:,i] = -1.0*E_V[:,:,i]

    E_flux_rad[:,i] = E_U[int(lenZ/delta)/2,x_center:x_center+int(Flux_rad/delta),i]
    Ez_flux[i] = np.sum(E_flux_rad[:,i])*2.0*np.pi
    Ez_center[i] = E_flux_rad[0,i]
```

```python
skip = 15
fig, ax = plt.subplots(figsize=(80, 40))
q = ax.quiver(X[::skip,::skip], Z[::skip,::skip], E_V[::skip,::skip,10], E_U[::skip,::skip,10], units='xy', pivot='mid', cmap='magma')
ax.set_aspect('equal')
plt.xlabel('x-coordinate [1mm dot separation]')
plt.ylabel('z-coordinate [1mm dot separation]')
plt.show()
```

```python
figure = plt.figure(figsize=(20,10))
plt.plot(disk_separation, abs(Ez_flux))
plt.xlabel('Disk_separation [m]',fontsize=20)
plt.ylabel('Electric flux $\Phi_\mathrm{E}$ [V$\cdot$m] ',fontsize=20)
plt.title('{} [$\mu$m] flux radius, {} [m] plate radius, {} [m] hole radius '.format(Flux_rad*1e6, disk_radii, hole_radii), fontsize=25)
```

```python
figure = plt.figure(figsize=(20,10))
plt.plot(disk_separation, abs(Ez_center))
plt.xlabel('Disk separation [m]',fontsize=20)
plt.ylabel('Central electric field vector ($E\cdot\hat{z}$ [V/m]) ',fontsize=20)
plt.title('{} [$\mu$m] flux radius, {} [m] plate radius, {} [m] hold radii'.format(Flux_rad*1e6, disk_radii, hole_radii), fontsize=25)
```

So varying this hole up to this size seems pretty negligible. Keep in mind that this derivative appears to decrease quadratically so for larger hole sizes this could be of concern but we are more than likely will not go larger than a 6 mm diameter hole size. This 10 ppm figure could very well change as we vary the plate separation and diameter of the disk

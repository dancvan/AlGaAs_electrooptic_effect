```python
import numpy as np
import matplotlib.pyplot as plt
meas_dir = '/Users/daniel_vander-hyde/Documents/git/SU/algaas_electrooptic_effect/measurements/OLG/algaas/10_27_2020/95_mW_lock/excitations/'
plt.style.use('/Users/daniel_vander-hyde/Documents/git/my_python/matplotlib/stylelib/pptsize')
```

```python
exc = ['50mV', '75mV', '100mV', '125mV', '150mV', '200mV']
```

```python
initdb = []
initdeg = []
for i in range(len(exc)):
    initdb.append(np.array(np.zeros((300,2))))
    initdeg.append(np.array(np.zeros((300,2))))
```

```python
for i in range(len(exc)):
    initdb[i] = np.loadtxt(meas_dir + exc[i] + '/' + 'db.TXT')
    initdeg[i] = np.loadtxt(meas_dir + exc[i] + '/' + 'deg.TXT')
```

```python
plt.subplot(211)
for i in range(len(exc)):
    plt.semilogx(initdb[i][:,0],initdb[i][:,1], label = exc[i])
plt.legend()
plt.ylabel('dB')
plt.subplot(212)
for i in range(len(exc)):
    plt.semilogx(initdeg[i][:,0],initdeg[i][:,1], label = exc[i])
plt.legend()
plt.ylabel('phase [deg]')
plt.xlabel('freq [Hz]')    
```

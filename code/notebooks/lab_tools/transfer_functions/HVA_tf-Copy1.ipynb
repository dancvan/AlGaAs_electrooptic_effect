```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
folder = '../../../measurements/HVA_tf/02_19_2020/'
```

```python
amp_1 = np.loadtxt(folder + 'dB.TXT') 
phase_1 = np.loadtxt(folder + 'PH.TXT')
```

```python
fig = plt.figure(figsize=(17,15))
plt.subplot(211)
plt.title('High voltage amplifier with pomona box (see alog 412)',fontsize=25)
ax = plt.gca()
plt.ylabel('dB',fontsize=20)
plt.grid('True', which='both')
plt.grid('True')
ax.tick_params(which='both', width=1)
ax.tick_params(which='minor', length=2)
ax.tick_params(labelsize=15)
ax.yaxis.offsetText.set_fontsize(15)
ax.axhline(amp_1[0,1],color='red',label=str(max(amp_1[:,1])) +' dB')
plt.plot(amp_1[:,0],amp_1[:,1],label='half max amplitude')
plt.xlim([1,amp_1[-1,0]])
ax.set_xscale('log')
plt.legend(prop={'size':20})
plt.subplot(212)
plt.plot(phase_1[:,0],phase_1[:,1],label = 'half max amplitude')
ax = plt.gca()
plt.ylabel('phase (degrees)',fontsize=20)
plt.grid('True', which='both')
plt.grid('True')
ax.tick_params(which='both', width=1)
ax.tick_params(which='minor', length=2)
ax.tick_params(labelsize=15)
ax.yaxis.offsetText.set_fontsize(15)
plt.xlabel('frequency [Hz]',fontsize=20)
plt.xlim([1,phase_1[-1,0]])
ax.set_xscale('log')
plt.legend(prop={'size':20}, loc=3)
plt.savefig(folder + 'HVA_plus_pomona_tf.png',dpi=300,bbox_inches='tight')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('text', usetex=True)
folder ="../../measurements/HVA_PI/03_02_2020/"
```

```python
dbHVA1 = np.loadtxt(folder + 'db1.TXT')
phHVA1 = np.loadtxt(folder + 'deg1.TXT')
dbHVA2 = np.loadtxt(folder + 'db2.TXT')
phHVA2 = np.loadtxt(folder + 'deg2.TXT')
dbHVA3 = np.loadtxt(folder + 'db3.TXT')
phHVA3 = np.loadtxt(folder + 'deg3.TXT')
```

```python
f = dbHVA1[:,0]
amp1 = dbHVA1[:,1]
amp2 = dbHVA2[:,1]
amp3 = dbHVA3[:,1]
```

```python
fig = plt.figure(figsize=(30,20))
plt.semilogx(f,amp1,label='HVA 1')
plt.semilogx(f,amp2, label='HVA 2')
plt.semilogx(f,amp3, label= 'HVA 3')
plt.axvline(x=10e3,color='r',linewidth=5)
plt.xlim([f[0], f[-1]])
plt.legend(fontsize=30)
plt.xlabel('Frequency [Hz]',FontSize=30)
plt.ylabel('$$dB$$',FontSize=30)
plt.title("HVA Physik Instrumente", FontSize=50)
plt.xticks(fontsize=30, rotation=0)
plt.yticks(fontsize=30, rotation=0)
plt.grid(b=None, which='both', axis='both',linewidth=2)
ax = plt.axes()
ax.tick_params(axis='both', which='both',length=5,width=2)
plt.savefig(folder + 'HVA_PI.png',dpi=300,bbox_inches='tight')
```

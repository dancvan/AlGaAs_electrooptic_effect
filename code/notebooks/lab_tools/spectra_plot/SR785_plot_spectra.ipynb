# Now to calibrate into a laser noise spectra

```python
error_noise_at_10k = spectra[0][:,1][ind_int]
HVA_gain_at_10k = 1.07 #(V/V)
laser_detune_coeff = 1e6 #(Hz/V)

disp_noise_at_10k =error_noise_at_10k*HVA_gain_at_10k*laser_detune_coeff

print("Laser detune noise at 10 kHz is : {} Hz/sqrt(Hz)".format(disp_noise_at_10k[0]))
```

```python
title = 'FSR_12_06_2019'
data_in = np.genfromtxt(title + '.CSV',delimiter=',')
#data_out = np.genfromtxt('TEK00001.CSV',delimiter =',')
```

```python
plt.figure(figsize=(22,15))
plt.semilogy(time,ch3)
plt.xlabel('time (sec)')
```

```python
peaks = sig.find_peaks(ch3,height=4e-2,distance=100)
```

```python
def plot_peaksandlabels(time,voltage,peak_info):
    time_peak = time[peak_info[0]]
    voltage_peak = voltage[peak_info[0]]
    points = np.vstack((time_peak,voltage_peak)).T
    fig, ax = plt.subplots(figsize=(25,15))
    plt.plot(time,voltage)
    ax.scatter(time_peak, voltage_peak)
    for i, txt in enumerate(voltage_peak): 
        ax.annotate(txt, (time_peak[i], voltage_peak[i]),verticalalignment='bottom',horizontalalignment='center',fontsize=20)
    plt.xlabel('time [s]',fontsize=30)
    plt.ylabel('amplitude [V]', fontsize=30)
```

```python
plot_peaksandlabels(time, ch3, peaks)
plt.plot(time,ch4)
plt.title(title, fontsize=40)
plt.savefig(title + '_results.png',dpi=300,bbox_inches='tight')
```

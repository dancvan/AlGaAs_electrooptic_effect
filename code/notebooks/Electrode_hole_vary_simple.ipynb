```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
d = 2e-3
r_star = np.arange(0,10e-3,1e-4)
E = np.array(np.zeros(np.shape(r_star)))
R = 2.0*2.54e-2
for i in range(len(E)):
    E[i] = (d/(2.0*8.854e-12))*((1/np.sqrt(d**2.0 + r_star[i]**2.0)) -(1/np.sqrt(d**2.0 + R**2.0)))
```

```python
fig = plt.figure(figsize=(20,15))
plt.plot(r_star,E/(max(E)))
plt.xlim(r_star[0],r_star[-1])
ROI = 1.5e-3
plt.axvline(ROI,color='red')
```

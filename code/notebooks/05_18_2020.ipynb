```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('text', usetex=True)
folder ="../../measurements/debugging/05_18_2020/SCRN00"
```

```python
sr785_noise_floor = np.loadtxt(folder + '61.TXT')
```

```python
mixer_out_unplugged = np.loadtxt(folder + '60.TXT')
fast_mon_unplugged = np.loadtxt(folder + '62.TXT')
```

```python
mixer_out_pm = np.loadtxt(folder + '63.TXT')
fast_mon_pm = np.loadtxt(folder + '64.TXT')
fast_mon_longer_pm = np.loadtxt(folder + '65.TXT')
pd_rf_pm = np.loadtxt(folder + '66.TXT')
pd_dc_pm = np.loadtxt(folder + '67.TXT')
```

```python
mixer_out_om = np.loadtxt(folder + '70.TXT')
fast_mon_om = np.loadtxt(folder + '69.TXT')
pd_dc_om = np.loadtxt(folder + '68.TXT')
```

```python
ff = sr785_noise_floor[:,0]
```

```python
fig = plt.figure(figsize=(36.8,20))
plt.loglog(ff,sr785_noise_floor[:,1], label = 'SR785 noise floor')
plt.loglog(ff,mixer_out_unplugged[:,1], label = 'LO and PD unplugged')
plt.loglog(ff,mixer_out_om[:,1], label = 'Traditional mount + SiO2/TaO5 coated mirror')
plt.loglog(ff,mixer_out_pm[:,1], label = '3D printed mount + AlGaAs coated mirror')
plt.xlim(ff[0], ff[-1])
plt.ylim(1e-4, .4*1e-1)
plt.legend(fontsize=30)
plt.xlabel('Frequency [Hz]',FontSize=30)
plt.ylabel('$$V_{pk}$$',FontSize=30)
plt.title("Mixer out", FontSize=50)
plt.xticks(fontsize=30, rotation=0)
plt.yticks(fontsize=30, rotation=0)
plt.grid(b=None, which='both', axis='both',linewidth=2)
ax = plt.axes()
ax.tick_params(axis='both', which='both',length=5,width=2)
```

```python
fig = plt.figure(figsize=(36.8,20))
plt.loglog(ff,fast_mon_unplugged[:,1], label = 'LO and PD unplugged')
plt.loglog(ff,fast_mon_om[:,1], label = 'Traditional mount + SiO2/TaO5 coated mirror')
plt.loglog(ff,fast_mon_longer_pm[:,1], label = '3D printed mount + AlGaAs coated mirror')
plt.xlim(ff[0], ff[-1])
plt.legend(fontsize=30)
plt.xlabel('Frequency [Hz]',FontSize=30)
plt.ylabel('$$V_{pk}$$',FontSize=30)
plt.title("Fast Monitor", FontSize=50)
plt.xticks(fontsize=30, rotation=0)
plt.yticks(fontsize=30, rotation=0)
plt.grid(b=None, which='both', axis='both',linewidth=2)
ax = plt.axes()
ax.tick_params(axis='both', which='both',length=5,width=2)
```

```python
fig = plt.figure(figsize=(36.8,20))
plt.loglog(ff,pd_dc_om[:,1], label = 'Traditional mount + SiO2/TaO5 coated mirror')
plt.loglog(ff,pd_dc_pm[:,1], label = '3D printed mount + AlGaAs coated mirror')
plt.xlim(ff[0], ff[-1])
plt.legend(fontsize=30)
plt.xlabel('Frequency [Hz]',FontSize=30)
plt.ylabel('$$V_{pk}$$',FontSize=30)
plt.title("REFL PD (DC channel)", FontSize=50)
plt.xticks(fontsize=30, rotation=0)
plt.yticks(fontsize=30, rotation=0)
plt.grid(b=None, which='both', axis='both',linewidth=2)
ax = plt.axes()
ax.tick_params(axis='both', which='both',length=5,width=2)
```

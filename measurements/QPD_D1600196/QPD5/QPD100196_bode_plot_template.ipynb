## Amplitude and phase subplots (traditional bode)

```python
fig, axs = plt.subplots(2, figsize=(10,10))
axs[0].semilogx(pos_CH1.frequency[:], pos_CH1.amplitude[:],color='red',label='CH1 V+')
axs[0].semilogx(neg_CH1.frequency[:], neg_CH1.amplitude[:],'--',color='red', label='CH1 V-')
axs[0].semilogx(pos_CH2.frequency[:], pos_CH2.amplitude[:], color='blue', label='CH2 V+')
axs[0].semilogx(neg_CH2.frequency[:], neg_CH2.amplitude[:],'--', color='blue', label='CH2 V-')
axs[0].semilogx(pos_CH3.frequency[:], pos_CH3.amplitude[:],color='green',label='CH3 V+')
axs[0].semilogx(neg_CH3.frequency[:], neg_CH3.amplitude[:],'--',color='green', label='CH3 V-')
axs[0].semilogx(pos_CH4.frequency[:], pos_CH4.amplitude[:], color='magenta', label='CH4 V+')
axs[0].semilogx(neg_CH4.frequency[:], neg_CH4.amplitude[:],'--', color='magenta', label='CH4 V-')
axs[0].title.set_text('QPD5 V+ and V-')
axs[0].title.set_size(20)
axs[0].legend()

axs[0].set_ylabel('amplitude [dB]')
axs[1].semilogx(pos_ph_CH1.frequency[:], pos_ph_CH1.amplitude[:],color='red', label='CH1 V+')
axs[1].semilogx(neg_ph_CH1.frequency[:], neg_ph_CH1.amplitude[:],'--',color='red',label='CH1 V-')
axs[1].semilogx(pos_ph_CH2.frequency[:], pos_ph_CH2.amplitude[:],color='blue',label='CH2 V+')
axs[1].semilogx(neg_ph_CH2.frequency[:], neg_ph_CH2.amplitude[:],'--',color='blue',label='CH2 V-')
axs[1].semilogx(pos_ph_CH3.frequency[:], pos_ph_CH3.amplitude[:],color='green', label='CH3 V+')
axs[1].semilogx(neg_ph_CH3.frequency[:], neg_ph_CH3.amplitude[:],'--',color='green',label='CH3 V-')
axs[1].semilogx(pos_ph_CH4.frequency[:], pos_ph_CH4.amplitude[:],color='magenta',label='CH4 V+')
axs[1].semilogx(neg_ph_CH4.frequency[:], neg_ph_CH4.amplitude[:],'--',color='magenta',label='CH4 V-')
axs[1].legend()
axs[1].set_ylabel('phase [deg]')
axs[1].set_xlabel('frequency [Hz]')
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import sdfascii as sdf
```

```python
params = {"ytick.color" : "g",
          "xtick.color" : "g",
          "axes.labelcolor" : "g",
          "axes.edgecolor" : "g",
          "text.color" : "g",
         }
plt.rcParams.update(params)
```

## ASCII

```python
CH1_TP4_data = sdf.read_ascii_files('CH1_TP4/QPD5CH1T')
CH2_TP4_data = sdf.read_ascii_files('CH2_TP4/QPD5CH2T')
CH3_TP4_data = sdf.read_ascii_files('CH3_TP4/QPD5CH3T')
CH4_TP4_data = sdf.read_ascii_files('CH4_TP4/QPD5CH4T')
```

```python
fig = plt.figure(figsize=(20,15))
plt.semilogx(CH1_TP4_data.frequency[:], CH1_TP4_data.amplitude[:],label='CH 1')
plt.semilogx(CH2_TP4_data.frequency[:], CH2_TP4_data.amplitude[:],label='CH 2')
plt.semilogx(CH3_TP4_data.frequency[:], CH3_TP4_data.amplitude[:],label='CH 3')
plt.semilogx(CH4_TP4_data.frequency[:], CH4_TP4_data.amplitude[:],label='CH 4')
plt.xlabel('Frequency [Hz]', FontSize=20)
plt.ylabel('dB', FontSize=20)
plt.title('QPD5 TP4 transfer function measurements ($25 \mu$A injection current)', FontSize=25)
plt.legend()
```

```python
CH3_plus_amp = sdf.read_ascii_files('CH3_OUT+/QPD5C3R7')
CH3_neg_amp = sdf.read_ascii_files('CH3_OUT-/QPD5C3R9')
```

```python
fig = plt.figure(figsize=(20,15))
plt.semilogx(CH3_plus_amp.frequency[:], CH3_plus_amp.amplitude[:],label='Positive')
plt.semilogx(CH3_neg_amp.frequency[:], CH3_neg_amp.amplitude[:],label='Negative')
plt.xlabel('Frequency [Hz]', FontSize=20)
plt.ylabel('dB', FontSize=20)
plt.title('QPD5 TP4 transfer function measurements ($25 \mu$A injection current)', FontSize=25)
plt.legend()
```

```python
CH4_plus_amp = sdf.read_ascii_files('CH4_OUT+/QPD5R7')
CH4_plus_ph = sdf.read_ascii_files('CH4_OUT+/QPD5R7PH')
CH4_neg_amp = sdf.read_ascii_files('CH4_OUT-/QPD5R9')
```

```python
fig = plt.figure(figsize=(20,15))
plt.semilogx(CH4_plus_amp.frequency[:], CH4_plus_amp.amplitude[:],label='Positive')
plt.semilogx(CH4_neg_amp.frequency[:], CH4_neg_amp.amplitude[:],label='Negative')
plt.xlabel('Frequency [Hz]', FontSize=20)
plt.ylabel('dB', FontSize=20)
plt.title('QPD5 TP4 transfer function measurements ($25 \mu$A injection current)', FontSize=25)
plt.legend()
```

```python
fig = plt.figure(figsize=(20,15))
plt.semilogx(CH4_plus_amp.frequency[:], CH4_plus_ph.amplitude[:],label='Positive')
```

## SDF

```python
log_scale_meas = sdf.read_sdf_file('FLOOR.DAT')
lin_scale_meas_avg_30 = sdf.read_sdf_file('FLOOR30.DAT')
#tbl1 = sdf.read_sdf_file('DATA1.TBL') 
#sv_state= sdf.read_sdf_file('STAT1.STA')
```

```python
fig = plt.figure(figsize=(20,15))
freq_1 = np.arange(0, len(log_scale_meas[1])*128,128)
freq_2 = np.arange(0,len(lin_scale_meas_avg_30[1])*32,32)
plt.plot(freq_1, log_scale_meas[1], label='First measurement (400 line resolution)')
plt.plot(freq_2, lin_scale_meas_avg_30[1], label='Second measurement (1600 line resolution)')
plt.xlabel('Frequency (Hz)', FontSize=20)
plt.ylabel('What are these units?', FontSize=20)
plt.title('HP35670A dark noise measurement (SDF data)', FontSize=25)
plt.legend(prop={'size' : 15})
plt.ylim((20e-8, 1e-6))
```
